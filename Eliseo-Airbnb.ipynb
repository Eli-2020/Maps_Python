Finally, two more columns were added, the first one to group the prices into categories based on the first price charts, this one is called price_label; their ranges are made with the pd.cut function where the numbers are given by creating five categories and replacing the previous value in the column with the corresponding category. In the second new column, something similar is done but with the idea of assigning categories with names so that you can make visible which are the areas that have the categories with the highest prices that are: 4 and 5 by density color.

```python
airbnb["price_category"]=airbnb["price"]
airbnb.columns
```

```python
airbnb.dtypes
```

```python
airbnb["price_category"] = airbnb["price_category"].astype('float64')
```

```python
airbnb.dtypes
```

```python
bins=[0,100,175,250,500,10000]
names=["1","2","3","4","5"]
airbnb["price_category"]=pd.cut(airbnb["price_category"],bins,labels=names)
```

```python
plt.figure(figsize=(8, 7))
sns.scatterplot(x=airbnb['longitude'], y=airbnb['latitude'], hue=airbnb['price_category'], size=airbnb['price_category'])
plt.title('Category of Prices in NY by color density')
```

```python
airbnb["price_label"]=airbnb["price"]
airbnb.columns
```

```python
bins=[0,100,175,250,500,10000]
names=["Low","Medium","High","Businness","Monarchy"]
airbnb["price_label"]=pd.cut(airbnb["price_label"],bins,labels=names)
```

```python
plt.figure(figsize=(8, 7))
sns.scatterplot(x=airbnb['longitude'], y=airbnb['latitude'], hue=airbnb['price_label'])
plt.title('Category of Prices in NY')
```
